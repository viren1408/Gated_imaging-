```python
import datetime
import math
for mjd_obs in observation_mjds_values:
    # Given values
    ra_str = "08h35m20.5554429s" #RA of Pulsar 
    afmjd = mjd_obs  # mjd1_gmrt_round_new    #60268 #60297  # Example value for afmjd(MJD_start)
    maxha = 12 #  maxha 
    #tspan = 60  #  tspan (min)

    # Observatory coordinates  1656342.30    5797947.77 
    obs_x =  1656342.30  # Replace with the actual x-coordinate
    obs_y = 5797947.77 # Replace with the actual y-coordinate

    # Convert transit LST to datetime
    #transit_lst_time = datetime.datetime.strptime(transit_lst_str, "%H:%M:%S").time() 60312.97916666660

    # Convert RA to hours
    ra_hours = (8 + (35/60) + (20.5554429/3600))

    # Conversion of UT into local sidereal time
    solsid = 1.002737909

    #  LST transit in hours (taken from the NCRA transit time calculator)
    #hlst_gmrt = transit_lst_time.hour + transit_lst_time.minute/60 + transit_lst_time.second/3600 

    # Calculate observatory longitude
    alng = math.atan2(-obs_y, obs_x)

    # Conversion of UT into local sidereal time
    solsid = 1.002737909
    hlst = 24.0 * ((solsid * afmjd + 0.154374 - alng / (2 * math.pi)) % 1.0)

    
    rax = int(ra_hours) + int((ra_hours - int(ra_hours)) * 60.0 + 2) / 60.0

    wait_gmrt = (rax - hlst ) / solsid #GMRT # Calculate wait : Local hour angle

    if wait_gmrt < -maxha:
        wait_gmrt += 24.0/solsid

    print(f"TMID:{mjd_obs} HA:{wait_gmrt}")
```

 Here it can be seen that tempo2 has added almost a day to understand why this was done we check the local HA at the start MJD : 60314 
 On observation: 
   1. In tempo2 the range of MJD at which the phases will be calculated is decided by a formula given in the report. 
   2. There we can see that it creates an offset and that offset decides the range in which polyco outputs are generated 

```python
# Given values
import math
ra_str = "08h35m20.5554429s" #RA of Pulsar 
afmjd = 60314 # mjd1_gmrt_round_new    #60268 #60297  # Example value for afmjd(MJD_start)
maxha = 12 #  maxha 
tspan = 60  #  tspan (min)

# Observatory coordinates  1656342.30    5797947.77 
obs_x =  1656342.30  # Replace with the actual x-coordinate
obs_y = 5797947.77 # Replace with the actual y-coordinate

# Convert transit LST to datetime
#transit_lst_time = datetime.datetime.strptime(transit_lst_str, "%H:%M:%S").time() 60312.97916666660

# Convert RA to hours
ra_hours = (8 + (35/60) + (20.5554429/3600))

# Conversion of UT into local sidereal time
solsid = 1.002737909

#  LST transit in hours (taken from the NCRA transit time calculator)
#hlst_gmrt = transit_lst_time.hour + transit_lst_time.minute/60 + transit_lst_time.second/3600 

# Calculate observatory longitude
alng = math.atan2(-obs_y, obs_x)

# Conversion of UT into local sidereal time
solsid = 1.002737909
hlst = 24.0 * ((solsid * afmjd + 0.154374 - alng / (2 * math.pi)) % 1.0)


rax = int(ra_hours) + int((ra_hours - int(ra_hours)) * 60.0 + 2) / 60.0

wait_gmrt = (rax - hlst ) / solsid #GMRT # Calculate wait : Local hour angle

if wait_gmrt < -maxha:
    wait_gmrt += 24.0/solsid

wait_gmrt
offset = ((wait_gmrt - maxha) / float(24.0)) + (tspan / float(2 * 1440.0))
print(f"The offset for MJD_start {afmjd} adjusted by polyco is {offset} and HA is {wait_gmrt}")
```
