```python
def calculate_rotations(tspan_minutes, frequency_hz):
    # Convert time span from minutes to seconds
    tspan_seconds = tspan_minutes * 60
    
    # Calculate the number of rotations using the formula: rotations = time span * frequency
    rotations = tspan_seconds * frequency_hz
    
    return rotations

# Example usage:
tspan = 120 # time span in minutes
f0 = 11.188233045354 # frequency in Hertz

result = calculate_rotations(tspan, f0)
print(f"Number of rotations in {tspan} minutes at {f0} Hz: {result}")
```
